# GPS

```python
#read in the dataset
gps = pd.read_csv("gps.csv")
#change tiemstamp type to datetime
gps["Timestamp"] = gps["Timestamp"].apply(lambda x: dt.datetime.strptime(x,"%m/%d/%Y %H:%M:%S"))
#string for graphs to display as category
#may not need this later
# gps['idStr']=gps['id'].astype(str)

#manipulating timestamp
def hour_converter(x):
    if (x >= 3) and (x <= 11):
        return 'Morning'
    elif (x >= 11) and (x <= 16):
        return'Lunch'
    elif (x >= 16) and (x <= 20):
        return'Evening'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x >= 0) and (x <= 3):
        return'MidNight'
    elif (x > 3) and (x < 6):
        return 'Early Morning'
gps['hour'] = pd.to_datetime(gps['Timestamp']).dt.hour
gps['time'] = gps['hour'].apply(hour_converter)
gps['date'] = gps['Timestamp'].dt.date
gps["is_weekend"] = gps['Timestamp'].dt.dayofweek > 4
gps.head()
```

```python
#what does this look like give me some info
print(gps.shape)
gps.info()

columns=list(gps.columns)
for col in columns:
    print(col, ": ")
    print(gps[col].unique())
    print()
```

# Merging GPS and Car Assignments

```python
#merging the car_assignment dataset
gps = pd.merge(gps, car_assignments, left_on='id', right_on= 'CarID', how='outer')#
#taking out the ones that couldn't be merged (truck drivers)
gps=gps[gps['Timestamp'].notna()]
#dropping carid column
gps.drop(['CarID'], axis=1,inplace=True)

#might be helpful later
# gps.loc[(gps['id']==101) & (gps['date']=='2014-01-06'), 'FullName']="Albina Hafon"
gps
```

```python
#filling in the nulls for car id for truck drivers
def nullTD(row):
    if (row['id']==101):
        albina=[dt.datetime.strptime(date, '%Y-%m-%d').date() for date in ['2014-01-06', '2014-01-07', '2014-01-13']]
        benito= [dt.datetime.strptime(date, '%Y-%m-%d').date() for date in ['2014-01-09', '2014-01-14', '2014-01-16']]
        claudio=pd.Timestamp('2014-01-10')
        if row['date'] in albina:
            return 'Albina Hafon'
        if row['date'] in benito:
            return 'Benito Hawelon'
        if row['date']==dt.datetime.strptime('2014-01-10', '%Y-%m-%d'):
            return 'Claudio Hawelon'
    if (row['id']==106):
        dylan= [dt.datetime.strptime(date, '%Y-%m-%d').date() for date in ['2014-01-07', '2014-01-09', '2014-01-14', '2014-01-16']]
        adan= [dt.datetime.strptime(date, '%Y-%m-%d').date() for date in ['2014-01-08', '2014-01-15']]
        if row['date'] in dylan:
            return 'Dylan Scozzese'
        if row['date'] in adan:
            return 'Adan Morlun'
    
    if (row['id']==107):
        cecilia= [dt.datetime.strptime(date, '%Y-%m-%d').date() for date in ['2014-01-06', '2014-01-13', '2014-01-14', '2014-01-16', '2014-01-17']]
        if row['date'] in cecilia:
            return 'Cecilia Morluniau'
        else:
            if row['hour'] <12:
                return 'Cecilia Morluniau'
            else:
                return 'Irene Nant'
        
gps['FullName'][gps['FullName'].isnull()]=gps.apply(lambda row: nullTD(row), axis=1)
gps['CurrentEmploymentType'].fillna("Facilities", inplace=True)
gps['CurrentEmploymentTitle'].fillna("Truck Driver", inplace=True)
```

### Stop Maker

```python
#making the id list
idList=np.sort(gps['id'].unique())
```

```python
#making stops
for d in idList:
    data=gps[gps['id']==d].sort_values(['Timestamp'])
    
    stopList=[]
    
    for i in range(len(data['Timestamp'])-1):
        point=data['Timestamp'].iloc[i]
        nextPoint=data['Timestamp'].iloc[i+1]
    #     print(type(nextPoint-point))
        if (nextPoint-point) > dt.timedelta(minutes=3):
    #         print(nextPoint-point)
            stopList.append(1)
        else:
            stopList.append(0)

    stopList.append(0)
#     print(stopList)
    data['stop']=stopList
    if d==1:
        gpsFull=data.copy()
        dataStops=data[data['stop']==1].copy()
    else:
        gpsF=data.copy()
        gpsFull=pd.concat([gpsFull, gpsF], ignore_index=True,axis=0)
        dataS=data[data['stop']==1].copy()
        dataStops=pd.concat([dataStops, dataS], ignore_index=True,axis=0)
dataStops
```

```python
#fixing the file
cc_data["timestamp"] = cc_data["timestamp"].apply(lambda x: dt.datetime.strptime(x,"%m/%d/%Y %H:%M"))
cc_data["timestampFull"]=cc_data["timestamp"].copy()
cc_data['timestamp'] = cc_data['timestamp'].dt.strftime('%m/%d/%Y')
cc_data["timestamp"] =  pd.to_datetime(cc_data["timestamp"], infer_datetime_format=True)
# make a cc binary col
cc_data['cc']=1
cc_data['cents']= cc_data['price'].apply(lambda x: round(x - np.fix(x),2))
```
